```python
import pandas as pd
import numpy as np
from pycaret.datasets import get_data
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer 

import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_excel('bank-additional-full-cleaned.xlsx')
df.head()
```

```python
df.info()
```

Berdasarkan Inferensial di EDA kolom housing dan loan tidak memiliki pengaruh signifikan

```python
df = df.drop(columns='housing')
df = df.drop(columns='loan')
```

---

```python
from pycaret.classification import *
s = setup(df, target = 'deposit', session_id = 42,normalize=True,encoding_method='one_hot',max_encoding_ohe= 25,normalize_method='robust', categorical_imputation= 'mode',fix_imbalance=True,fix_imbalance_method='randomoversampler')
```

```python
get_config('pipeline')
```

```python
# functional API
best = compare_models()

# OOP API
best = s.compare_models()
```

```python
# functional API
predict_model(best)

# OOP API
s.predict_model(best)
```
